# 1. Import necessary libraries

```python

from sklearn.datasets import load_iris #Built in Iris data set
from sklearn.model_selection import train_test_split # This model splits data into two sets for training the model and testing it.
from sklearn.tree import DecisionTreeClassifier #This is the model we going to use to train
from sklearn.metrics import accuracy_score, precision_score, recall_score # Eveluates our model
from sklearn.preprocessing import StandardScaler # process out data into easy digestible units that a machin can read.
```

# 2. Load Iris dataset

```python

iris = load_iris()
X, y = iris.data, iris.target
```

# 3. Preprocess (optional for tree models, but for demonstration we standardize features)

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# 4. Split data into training and testing sets



```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

# 5. Train a Decision Tree Classifier

```python

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

# 6. Predict on test data


```python
y_pred = model.predict(X_test)
```

# 7. Evaluate using accuracy, precision, and recall


```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')  # macro for multi-class
recall = recall_score(y_test, y_pred, average='macro')
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

# Viuslaization of Accuracy,preciosion and Recall

```python
import matplotlib.pyplot as plt
```

```python
# Metrics
metrics = ['Accuracy', 'Precision', 'Recall']
values = [accuracy, precision, recall]

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, values, color=['skyblue', 'lightgreen', 'salmon'])

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f"{yval:.2f}", ha='center', va='bottom')

plt.ylim(0, 1.1)
plt.title('Decision Tree Classifier Performance')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
